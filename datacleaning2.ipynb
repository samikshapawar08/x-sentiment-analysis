```python
import pandas as pd
import re
```

```python
df = pd.read_csv('Twitter_Data.csv')
```

```python
df.dropna(inplace=True)
```

```python
def clean_text(text):
    text = str(text).lower()  # lowercase
    text = re.sub(r"http\S+|www\S+", "", text)  # remove URLs
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # remove special characters
    return text.strip()
```

```python
df['cleaned_text'] = df['clean_text'].apply(clean_text)
```

```python
df['Sentiment'] = df['category'].map({
    1: "Positive",
    0: "Neutral",
    -1: "Negative"
})
```

```python
df['sentiment_score'] = df['Sentiment'].map({
    "Positive": 1,
    "Neutral": 0,
    "Negative": -1
})
```

```python
df['word_count'] = df['cleaned_text'].apply(lambda x: len(x.split()))
```

```python
df['text_length'] = df['cleaned_text'].apply(len)
```

```python
keywords = ['issue', 'problem', 'delay', 'bad', 'error', 'fail']

def check_issue(text):
    for word in keywords:
        if word in text:
            return 1
    return 0

df['has_issue'] = df['cleaned_text'].apply(check_issue)
```

```python
df = df[['clean_text', 'cleaned_text', 'Sentiment', 
         'sentiment_score', 'word_count', 'text_length', 'has_issue']]
```

```python
df.to_csv('cleaned_twitter_data.csv', index=False)

print("✅ Cleaned data with new columns saved successfully!")
```
