```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('cleaned_twitter_data.csv')

# Combine text
text = " ".join(df['cleaned_text'].dropna())

# Stopwords
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(["amp", "rt"])

# Function to make all words white
def white_color_func(*args, **kwargs):
    return "white"

# Create word cloud
wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color=None,   # transparent
    mode='RGBA',             # needed for transparency
    stopwords=custom_stopwords,
    max_words=150,
    color_func=white_color_func  # 👈 IMPORTANT
).generate(text)


wordcloud.to_file("wordcloud_white.png")


plt.imshow(wordcloud)
plt.axis('off')
plt.show()

print("✅ White word cloud saved!")
```
